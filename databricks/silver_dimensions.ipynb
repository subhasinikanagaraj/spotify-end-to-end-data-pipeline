```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
from datetime import datetime
```

```python


def dcp(tbl):
    ts = datetime.now().strftime("%Y%m%d_%H%M%S")
    dcp = f"abfss://silver@dlspotifyde.dfs.core.windows.net/display_checkpoint/{tbl}/checkpoint_{ts}"
    return dcp
```

# DimUser

```python
df = spark.read.format("parquet").load('abfss://bronze@dlspotifyde.dfs.core.windows.net/DimUser')
df.display()
```

## AutoLoader

```python
df_user = spark.readStream.format("cloudFiles")\
                .option("cloudFiles.format","parquet")\
                .option("cloudFiles.schemaLocation","abfss://silver@dlspotifyde.dfs.core.windows.net/DimUser/stream_checkpoint")\
                .option("schemaEvolutionMode","addNewColumns")\
                .load("abfss://bronze@dlspotifyde.dfs.core.windows.net/DimUser")
df_user.display(checkpointLocation=dcp("DimUser"))
```

```python
df_user = df_user.withColumn("user_name" , upper(col("user_name")))
df_user.display(checkpointLocation=dcp("DimUser"))
```

```python
df_user = df_user.drop('_rescued_data')
#df_user = df_user.dropDuplicates(['user_id'])
display(df_user,checkpointLocation=dcp("DimUser"))
```

```python
df_user.writeStream.format("delta")\
        .outputMode("append")\
        .option("checkpointLocation","abfss://silver@dlspotifyde.dfs.core.windows.net/DimUser/stream_checkpoint")\
        .trigger(once=True)\
        .option("path","abfss://silver@dlspotifyde.dfs.core.windows.net/DimUser/data")\
        .toTable("spotify.silver.DimUser")
```

# DimTrack

```python
df_track = spark.readStream.format("cloudFiles")\
                .option("cloudFiles.format","parquet")\
                .option("cloudFiles.schemaLocation","abfss://silver@dlspotifyde.dfs.core.windows.net/DimTrack/stream_checkpoint")\
                .option("schemaEvolutionMode","addNewColumns")\
                .load("abfss://bronze@dlspotifyde.dfs.core.windows.net/DimTrack")
df_track.display(checkpointLocation=dcp("DimTrack"))
```

```python
df_track = df_track.withColumn("duration_catgory",   when(col('duration_sec')<=150, "short")\
                                                    .when(col('duration_sec')<=300, "medium")\
                                                    .otherwise("long")\
                                )
df_track = df_track.withColumn("track_name", regexp_replace(col('track_name'),'-',' '))
df_track = df_track.drop('_rescued_data')
#df_track = df_track.dropDuplicates(['track_id'])
display(df_track,checkpointLocation=dcp("DimTrack"))
```

```python
df_track.writeStream.format("delta")\
        .outputMode("append")\
        .option("checkpointLocation","abfss://silver@dlspotifyde.dfs.core.windows.net/DimTrack/stream_checkpoint")\
        .trigger(once=True)\
        .option("path","abfss://silver@dlspotifyde.dfs.core.windows.net/DimTrack/data")\
        .toTable("spotify.silver.DimTrack")
```
